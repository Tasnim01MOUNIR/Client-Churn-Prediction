```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

# Channel sales

```python
# Select the 'id', 'channel_sales', and 'churn' columns from the client_df DataFrame
channel = client_df[['id', 'channel_sales', 'churn']]

# Group the data by 'channel_sales' and 'churn', and count the number of 'id' values in each group
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count()

# Unstack the grouped data, making 'churn' values (0 and 1) into separate columns
channel = channel.unstack(level=1)

# Fill any missing (NaN) values with 0
channel = channel.fillna(0)

# Calculate the percentage of churn (1) for each 'channel_sales' category
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100)

# Sort the resulting DataFrame by the percentage of churn (1) in descending order
channel_churn = channel_churn.sort_values(by=[1], ascending=False)
```

```python
plot_stacked_bars(channel_churn, 'Sales channel', rot_=30)
```

Interestingly, the churning customers are distributed over 5 different values for channel_sales

# Consumption 

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(10, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```


The consumption data exhibits a pronounced positive skewness, 
indicating that it is skewed towards higher values with an elongated right tail in its distribution.
This elongation suggests the presence of outliers, which are values that significantly deviate from the majority of the data points.
To gain deeper insights into these outliers, we can employ a standard visualization technique – the boxplot.

A boxplot offers a standardized representation of the data's distribution by highlighting key statistical measures:
- The minimum value
- The first quartile (Q1)
- The median
- The third quartile (Q3)
- The maximum value

This visualization not only helps us identify outliers but also provides information about the data's symmetry, 
the tightness of data clustering, and the extent of its skewness. It is a valuable tool for gaining a comprehensive understanding of the data's characteristics.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the columns you want to create boxplots for
columns_to_plot = ["cons_12m", "cons_gas_12m", "cons_last_month", "imp_cons"]

# Create a figure with subplots
fig, axs = plt.subplots(nrows=len(columns_to_plot), figsize=(11, 4 * len(columns_to_plot)))

# Loop through columns and create boxplots with customizations
for i, column in enumerate(columns_to_plot):
    sns.boxplot(x=column, data=consumption, ax=axs[i], linewidth=1, notch=True, fliersize=5, palette="Set3")
    axs[i].set_title(f"Boxplot of {column}")
    axs[i].set_xlabel(column)

plt.tight_layout()  # Adjust subplot spacing
plt.show()
```

There are  a lot of outliers
